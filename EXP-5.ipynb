1. Smoothing Filters

a) Using Averaging Filter

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("cat39.jpeg")
image1=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image1)
plt.axis("off")
```

```python
avg_blur = cv2.blur(image1,(11,11))

plt.figure(figsize=(9,9))

plt.subplot(1,2,1)
plt.imshow(image1)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(avg_blur)
plt.title("Averaging Filter")
plt.axis("off")

plt.show()
```

b) Using Weighted Averaging Filter

```python
kernel = np.array([[1,2,1],[2,4,2],[1,2,1]])/16
weighted_average_filter = cv2.filter2D(image, -1, kernel)
```

```python
plt.figure(figsize=(9,9))

plt.subplot(121)
plt.imshow(image[:,:,::-1])
plt.title("Input Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(weighted_average_filter[:,:,::-1])
plt.title("Output Image (Weighted Average Filter)")
plt.axis("off")

plt.show()
```

c) Using Gaussian Filter

```python
gaussian_blur=cv2.GaussianBlur(image1,(33,33),0,0)
plt.figure(figsize=(9,9))

plt.subplot(1,2,1)
plt.imshow(image1)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(gaussian_blur)
plt.title("Gaussian Blur")
plt.axis("off")
plt.show()
```

d) Using Median Filter

```python
median=cv2.medianBlur(image1,13)
plt.figure(figsize=(9,9))
plt.subplot(1,2,1)
plt.imshow(image1)

plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(median)
plt.title("Median Blur")
plt.axis("off")
plt.show()
```

2. Sharpening Filters

a) Using Laplacian Linear Kernal

```python
lap_kernel = np.array([[0, -1, 0],[-1, 5, -1],[0, -1, 0]])
sharpened_lap_kernel = cv2.filter2D(image, -1, kernel = lap_kernel)

plt.figure(figsize=(9,9))

plt.subplot(121)
plt.imshow(image1)
plt.title("Input Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(sharpened_lap_kernel[:,:,::-1])
plt.title("Output Image (Laplacian Filter)")
plt.axis("off")

plt.show()
```

b) Using Laplacian Operator

```python
gray = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
lap = cv2.Laplacian(gray,cv2.CV_8U)

plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.imshow(image1)
plt.title("Input Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gray,cmap='gray')
plt.title("Gray_image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(lap,cmap='gray')
plt.title("Output Image (Laplacian_filter)")
plt.axis("off")

plt.show()
```
